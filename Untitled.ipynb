```python
seq_len_list = []

for u in u_list:
    df_u = df[df["userSeq"] == u]
    c5_list = np.unique(df_u["category3"].values)
    
    for c5 in c5_list:
        df_u_c5 = df_u[df_u["category3"] == c5]
        cunit_list = np.unique(df_u_c5["cunitSeq"].values)
        
        seq_len = cunit_list.shape[0]
        if seq_len > 1:
            seq_len_list.append(seq_len)

print(sum(seq_len_list), len(seq_len_list))

import matplotlib.pyplot as plt

plt.figure(figsize=[15, 10], facecolor="white")
plt.hist(seq_len_list, bins=100)
plt.show()
```

```python
df_ = df.copy()
df_["date"] = [row["regdate"][:10] for _, row in df.iterrows()]
```

```python
seq_len_list = []

for u in u_list:
    df_u = df_[df_["userSeq"] == u]
    c5_list = np.unique(df_u["category3"].values)
    
    for c5 in c5_list:
        df_u_c5 = df_u[df_u["category3"] == c5]
        cunit_list = np.unique(df_u_c5["date"].values)
        
        seq_len = cunit_list.shape[0]
        if seq_len > 1:
            seq_len_list.append(seq_len)

print(sum(seq_len_list), len(seq_len_list))

import matplotlib.pyplot as plt

plt.figure(figsize=[15, 10], facecolor="white")
plt.hist(seq_len_list, bins=100)
plt.show()
```
